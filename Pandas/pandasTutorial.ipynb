Data stored in pandas can be accessed and edited much faster than a python list for many reasons, but one reason behind this is pandas allows vectorizing operations where you can apply a function to an entire series as a whole in 1 sweep as opposed to iterating over each element in a list and applying a function one-by-one.  Similar to pandas, numpy also increases performance as discussed at [this link](https://towardsdatascience.com/speed-testing-pandas-vs-numpy-ffbf80070ee7) on towardsdatascience. The 2 main structures of pandas include a series and a dataframe. A series is 1 individual column often shorthanded as `s` while a dataframe is a set of series often shorthanded as `df`.

```python
import pandas as pd
import numpy as np
```

## Getting Started
To get started, you can create a dataframe using built-in pandas methods like `DataFrame()` and passing a dictionary with it's values as a list. Note that the length of the lists must match or you'll throw a `ValueError`.

```python
df = pd.DataFrame({'col1':[1,2,3], 'col2':[4,5,6], 'col3':[7,8,9]})
df
```

More commonly, you can import spreadsheets as dataframes using `read_excel()` or `read_csv()`. If you want to return the top or bottom *n* rows of your dataset, you can use `df.head(n)` or `df.tail(n)`, respectively with a default *n* = 5.

```python
df = pd.read_csv('avocado.csv')
df.head()
```

Show the series names and the associated types using `df.dtypes`.

```python
df.dtypes
```

One thing to note about the dtypes is that the dtype object can be strings or a mix of different types.  In this case, the `Date` series read in as a string from the csv import, so we can cast that series to it's appropraite dtype using `s.astype()`. Alternatively, you can use `df.astype(d)` where `d` is a dictionary map of the series names and types you wish to cast.

```python
df['Date'] = df['Date'].astype('datetime64')
df.dtypes
```

If you're only interested in the shape, you can use `df.shape` to return a tuple containing the matrix shape in the form of (*rows*, *columns*). This is useful when making sure the shape of two large dataframes match when manual checking isn't a feasible option.

```python
df.shape
```

Series are accessed by passing the series's name in `[]` or a list of names.

```python
df['Date']
```

```python
df[['Date', 'Total Volume', 'AveragePrice']].head()
```

You can also pass in a boolean to access datapoints that would return `True` for a statement.

```python
df[df['Date'] > '2018-01-01']
```

2 useful methods to see basic metrics for a dataframe or series is by calling `df.describe()` for the statistics on the numerical fields, or you use `df.nunique()` to see how many unique options are available in each series.

```python
df.describe()
```

```python
df.nunique()
```

If you want to see statistics for a given value in the data use the `df.groupby()` followed by what type of calculation you are interested in.  You can also sort the dataframe using `df.sort_values()`. In the event of a tie, you can pass multiple series names in the `by=[]` argument and pandas will use the subsequent values as tie breakers.

```python
df.groupby(by='region').mean()\
    .sort_values(by=['AveragePrice'], ascending=False)
```

## Data Editing

By default, pandas returns a view of the original dataframe or series. While it's possible to edit the original dataframe through a view, if you need to do some edits or calculations, it's best practice to return a copy using `df.copy()` to avoid a `SettingWithCopyWarning` - otherwise, you can run into issues later down the pipeline.

```python
df_select = df[['Date', 'AveragePrice', 'Total Volume', 'Total Bags', 'type', 'year', 'region']].copy()
df_select.head()
```

If the names of the series need cleaning or renaming entirely, use `df.rename()` and pass a dictionary of the old and new columns in the `columns` parameter. This method also returns a view, so to keep the changes permanent, you can either define a new variable or pass `inplace=True`. One thing to note is that if the key you are trying to rename isn't present in the dataframe, pandas *won't* throw an error as seen below with the non-existent `rotten` column.

```python
df_select.rename(columns={'AveragePrice':'Average Price USD', 'type':'Type', 'year':'Year', 'region':'Region', 'rotten': 'Rotten'}, inplace=True)
df_select.head()
```

## Performance

Applying a function to a series can be done using the `s.apply(f)` where `f` is a function. To make a new series, you can treat it like it the same way as editing an existing series. 

As mentioned earlier, a significant advantage pandas has over base python is vectorizing operations.  The below cells run the same operation and compares the time it takes to compute for the dataset.

```python
### Non-vectorized ###
%timeit df_select['Average Price MXN'] = df_select['Average Price USD'].apply(lambda x: round(x * 20.17, 2))
df_select.head()
```

```python
### Vectorized ###
%timeit df_select['Average Price CAD'] = round(df_select['Average Price USD'] * 1.3, 2)
df_select.head()
```

The chart below shows the computational speeds for various methods.

![speed chart](https://i.stack.imgur.com/sbtDB.png)

This shows calcuations using python lists follows a $O(NlogN)$ time complexity while vectorized operations follows $O(N)$.  For for information on Big O Notation, see [this article](https://towardsdatascience.com/introduction-to-big-o-notation-820d2e25d3fd) on towardsdatascience.

For a quick exploration of Tampa's organic avocado sales, you can create a dataframe filtering the region and type passing each boolean in seperate `()`, sort the dataframe by the `Date` series, and reset the index to date.

```python
df_tampa = df_select[(df_select['Region'] == 'Tampa') & (df_select['Type'] == 'organic')]\
    .sort_values(by=['Date'])\
    .set_index('Date')
df_tampa.head()
```

Pandas has built-in windows functions using `df.rolling()` or `df.expanding()`. This can be useful for calcualting the rolling average during a set number of periods or cumulative periods for an alternative to exponential smoothing.

```python
df_tampa['Average Price USD'].plot()
df_tampa['Average Price USD'].rolling(3).mean().plot()
```
